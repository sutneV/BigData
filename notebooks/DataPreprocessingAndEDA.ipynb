```python
df['Height'] = df['Height'].round(2)
df['Weight'] = df['Weight'].round(1)
```

```python
# Convert the float variable to the nearest integer value
float_columns = ['Age', 'FCVC', 'NCP', 'CH2O', 'TUE', 'FAF']

for col in float_columns:
    # round to nearest whole number
    df[col] = round(df[col]).astype('int')
```

```python
# Create a BMI column
df['BMI'] = df['Weight'] / ((df['Height']) ** 2)
df['BMI'] = df['BMI'].round(1)
```

```python
# Create a new column 'Weight_Classification' based on BMI
def classify_obesity(BMI):
    if BMI < 18.5:
        return 'Underweight'
    elif 18.5 <= BMI <= 24.9:
        return 'Normal'
    elif 25.0 <= BMI <= 29.9:
        return 'Overweight'
    elif 30.0 <= BMI <= 34.9:
        return 'Obesity I'
    elif 35.0 <= BMI <= 39.9:
        return 'Obesity II'
    else:
        return 'Obesity III'

df['Weight_Classification'] = df['BMI'].apply(classify_obesity)

# Drop the old 'NObeyesdad' column
df.drop(columns=['NObeyesdad'], inplace=True)
```

```python
# Change ordinal to Categorical
df['FCVC'] = df['FCVC'].map({1: 'Never', 2: 'Sometimes', 3: 'Always'})
df['NCP'] = df['NCP'].map({1: 'Between 1 and 2', 2: 'Three', 3: 'More than 3'})
df['CH2O'] = df['CH2O'].map({1: 'Less than a liter', 2: '1 to 2 L', 3: 'More than 2 L'})
df['FAF'] = df['FAF'].map({0: 'Never', 1: '1-2x a week', 2: '2-3x a week', 3: '4-5x a week'})
df['TUE'] = df['TUE'].map({0: '0-2 hours', 1: '3-5 hours', 2: 'More than 5 hours'})

state_region_map = {
    'Wilayah Persekutuan Kuala Lumpur': 'Central',
    'Wilayah Persekutuan Putrajaya': 'Central',
    'Selangor': 'Central',
    'Negeri Sembilan': 'Central',
    'Melaka': 'Southern',
    'Johor': 'Southern',
    'Perak': 'Northern',
    'Penang': 'Northern',
    'Pulau Pinang': 'Northern',
    'Kedah': 'Northern',
    'Perlis': 'Northern',
    'Kelantan': 'East Coast',
    'Terengganu': 'East Coast',
    'Pahang': 'East Coast',
    'Sabah': 'East Malaysia',
    'Sarawak': 'East Malaysia'
}

# Map the 'State' column to 'Region'
df['State'] = df['State'].map(state_region_map)
```

```python
# Convert object/text variables to category variables
categorical_cols = [
    'Gender', 'family_history_with_overweight', 'FAVC', 'FCVC',
    'NCP', 'CAEC', 'SMOKE', 'CH2O', 'SCC', 'FAF', 'TUE', 'CALC',
    'MTRANS','State', 'Weight_Classification'
]

for col in categorical_cols:
    df[col] = df[col].astype('category')
```

```python
# Drop the NULL values and reset the index
df = df.dropna()
df.reset_index(drop=True, inplace=True)
df.info()
```

```python
# Create distribution plots similar to the reference image
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Weight Distribution (Blue theme like the reference)
ax1.hist(df['Weight'], bins=30, density=True, alpha=0.7, color='lightblue', edgecolor='blue', linewidth=0.5)
# Add KDE curve
from scipy import stats
weight_kde = stats.gaussian_kde(df['Weight'])
weight_x = np.linspace(df['Weight'].min(), df['Weight'].max(), 100)
ax1.plot(weight_x, weight_kde(weight_x), color='blue', linewidth=2)
ax1.set_title('Weight Distribution', fontsize=14, fontweight='bold')
ax1.set_xlabel('Weight (kg)', fontsize=12)
ax1.set_ylabel('Density', fontsize=12)
ax1.grid(True, alpha=0.3)

# Height Distribution (Red/Pink theme like the reference)
ax2.hist(df['Height'], bins=30, density=True, alpha=0.7, color='lightcoral', edgecolor='red', linewidth=0.5)
# Add KDE curve
height_kde = stats.gaussian_kde(df['Height'])
height_x = np.linspace(df['Height'].min(), df['Height'].max(), 100)
ax2.plot(height_x, height_kde(height_x), color='red', linewidth=2)
ax2.set_title('Height Distribution', fontsize=14, fontweight='bold')
ax2.set_xlabel('Height (m)', fontsize=12)
ax2.set_ylabel('Density', fontsize=12)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print some basic statistics
print("Weight Statistics:")
print(f"Mean: {df['Weight'].mean():.2f} kg")
print(f"Std: {df['Weight'].std():.2f} kg")
print(f"Range: {df['Weight'].min():.1f} - {df['Weight'].max():.1f} kg")
print()
print("Height Statistics:")
print(f"Mean: {df['Height'].mean():.2f} m")
print(f"Std: {df['Height'].std():.2f} m") 
print(f"Range: {df['Height'].min():.2f} - {df['Height'].max():.2f} m")
```

```python
# Plot bar plots for categorical variables
categorical_columns = ['Gender', 'family_history_with_overweight', 'FAVC',
                       'FCVC', 'NCP', 'CAEC', 'SMOKE', 'CH2O', 'SCC', 'FAF',
                       'TUE', 'CALC', 'MTRANS','State', 'Weight_Classification']

# Define color palette
colors = sns.color_palette('Set3')

# Define order
order_dict = {
    'Gender': ['Male', 'Female'],
    'FCVC': ['Never', 'Sometimes', 'Always'],
    'NCP': ['Between 1 and 2', 'Three', 'More than three'],
    'CAEC': ['no', 'Sometimes', 'Frequently', 'Always'],
    'SMOKE': ['yes', 'no'],
    'CH2O': ['Less than a liter', 'Between 1 and 2 L', 'More than 2 L'],
    'SCC': ['yes', 'no'],
    'FAF': ['Never', '1-2x a week', '2-3x a week', '4-5x a week'],
    'CALC': ['no', 'Sometimes', 'Frequently', 'Always'],
    'MTRANS': ['Walking', 'Bike', 'Public_Transportation', 'Motorbike', 'Automobile'],
    'State': ['Central', 'Southern', 'Northern', 'East Coast', 'East Malaysia'],
    'Weight_Classification': ['Underweight', 'Normal', 'Overweight',
                              'Obesity I', 'Obesity II', 'Obesity III']
}

plt.figure(figsize=(12, 23))
for i, col in enumerate(categorical_columns):
    plt.subplot(5, 3, i + 1)
    value_counts = df[col].value_counts()

    # Reorder the value counts if a specific order is defined
    if col in order_dict:
        value_counts = value_counts.reindex(order_dict[col])

    value_counts.plot(kind='bar', color=colors[:len(value_counts)], edgecolor='black')
    plt.title(col)
    plt.xlabel('')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Converting the Categorical variables to numerical
# Define the mapping dictionary
gender_mapping = {'Male': 0, 'Female': 1}
fcvc_mapping = {'Never': 0, 'Sometimes': 1, 'Always': 2}
ncp_mapping = {'Between 1 and 2': 1, 'Three': 2, 'More than 3': 3}
freq_mapping = {'no': 0, 'Sometimes': 1, 'Frequently': 2, 'Always': 3}
bin_mapping = {'yes': 1, 'no': 0}
ch2o_mapping = {'Less than a liter': 1, '1 to 2 L': 2, 'More than 2 L': 3}
faf_mapping = {'Never': 0, '1-2x a week': 1, '2-3x a week': 2, '4-5x a week': 3}
tue_mapping = {'0-2 hours': 1, '3-5 hours': 2, 'More than 5 hours': 3}
trans_mapping = {'Walking': 1, 'Bike': 2, 'Public_Transportation': 3,
                 'Motorbike': 4, 'Automobile': 5}
state_mapping = {'Central': 1, 'Southern': 2, 'Northern' : 3, 'East Coast' : 4, 'East Malaysia' :5 }
obes_mapping = {'Underweight': 1, 'Normal': 2, 'Overweight': 3, 'Obesity I': 4,
                'Obesity II': 5, 'Obesity III': 6}

# Convert the text values to numerical values using map()
df['Gender'] = df['Gender'].map(gender_mapping)
df['family_history_with_overweight'] = df['family_history_with_overweight'].map(bin_mapping)
df['FAVC'] = df['FAVC'].map(bin_mapping)
df['FCVC'] = df['FCVC'].map(fcvc_mapping)
df['NCP'] = df['NCP'].map(ncp_mapping)
df['CAEC'] = df['CAEC'].map(freq_mapping)
df['SMOKE'] = df['SMOKE'].map(bin_mapping)
df['CH2O'] = df['CH2O'].map(ch2o_mapping)
df['SCC'] = df['SCC'].map(bin_mapping)
df['FAF'] = df['FAF'].map(faf_mapping)
df['TUE'] = df['TUE'].map(tue_mapping)
df['CALC'] = df['CALC'].map(freq_mapping)
df['MTRANS'] = df['MTRANS'].map(trans_mapping)
df['State'] = df['State'].map(state_mapping)
df['Weight_Classification'] = df['Weight_Classification'].map(obes_mapping)
```

```python
# Convert Categorical to Integer and check the df structure
for col in df.select_dtypes(['category']).columns:
    df[col] = df[col].astype(int)

df.info()
```

```python
# Calculate correlation and create a Heat map with the numerical columns
# Drop Height and Weight columns as they are used in the BMI calculation for our target variable Weight_Classification. BMI to be dropped as well.
variables = df.drop(columns=['Height', 'Weight', 'BMI'])

# Calculation of the correlation matrix
corr_matrix = variables.corr()

# Creating Heat Maps
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```
